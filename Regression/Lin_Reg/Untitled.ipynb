```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')


train_data = train_data.dropna()
test_data = test_data.dropna()
```

```python
x = train_data['x']
y = train_data['y']
plt.scatter(x,y)
```

```python
x = np.asarray(x).reshape([699,1])
x_b = np.c_[np.ones((699,1)),x]
y = np.asarray(y).reshape([699,1])
```

```python
opt_w = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)
opt_w
```

```python
pred_y = x_b.dot(opt_w)
pred_y.shape
```

```python
plt.scatter(x,pred_y,color='r')
plt.scatter(x,y,s=1)
```

```python
x_test = test_data['x']
x_test = np.asarray(x_test).reshape((300,1))
x_test_b = np.c_[np.ones((300,1)),x_test]
```

```python
pred_test = x_test_b.dot(opt_w)
```

```python
plt.scatter(x_test,pred_test,color='r')
```
